```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False 
```

```python
ca=pd.read_csv('ca_list_copy.csv')
```

```python
# 对0值做异常值处理
ca=ca.replace(0.0,ca['zwyx'].mean())
```

```python
# 对zwyx做平均值统计，其他列做value_counts
print('职位月薪统计:',ca['zwyx'].mean())
ca_column=list(ca.columns)
ca_column.remove('Id')
ca_column.remove('zwyx')
a=ca['zwmc'].value_counts()
ca_df=pd.DataFrame(a)
ca_df.columns=['value']
i=0
for item in ca.loc[:,ca_column]:
    a=ca[item].value_counts()
    ca_df=pd.DataFrame(a)
    ca_df.columns=['value']
    df=pd.concat([df,ca_df])
print('其他数据统计')
df
```

```python
# 显示城市薪资Top10 
ca_info=ca.groupby(['dd'])['zwyx'].max().sort_values(ascending=False)[:10]
ca_info.plot()
plt.xlabel('城市')
plt.title(u'Top5薪资排名')
plt.legend(loc='best')
plt.show()
```
