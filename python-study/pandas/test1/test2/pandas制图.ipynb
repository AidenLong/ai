# 排序

```python
ser1 = pd.Series(np.arange(0,4),index = ['b','c','d','a'])
ser1
```

```python
ser1.sort_values()  #通过值进行排序
```

```python
ser1.sort_values(ascending=False)  #通过值排序倒序
```

```python
df1 = pd.DataFrame(np.random.randint(1,9,(4,4)),columns= ['b','c','a','d'],index = ['B','D','A','C'])
df1
```

```python
df1.sort_index()  #通过行索引排序
```

```python
df1.sort_index(axis = 1)  #通过列索引进行排序
```

```python
df1.sort_index(axis = 1,ascending=False)  #通过列索引进行排序
```

```python
df1.sort_values(by = ['d','b'])  #通过某列的值进行排序
```

# 练习

```python
'''
将附件中的数据导入DataFrame中，实现以下操作：
1，对异常数据（数值为0）进行值替换，替换为当前列的平均值；
2，对zwyx列的数据进行平均值统计，其他列做计数统计（提示使用value_counts()）；
3，得到zwmc字段的唯一值列表；
4，通过group函数，实现对于dd字段的分组，并按照城市计算每个城市的最大薪资，
    使用折线图，显示Top10城市。（选做）
'''

df = pd.read_csv('ca_list_copy.csv')
df.set_index('Id')
```

```python
#1，对异常数据（数值为0）进行值替换，替换为当前列的平均值；
#获取平均值
mean = df['zwyx'].mean()
#替换数据
data = df.replace({0.0:mean})
data
```

```python
#对zwyx列的数据进行平均值统计，其他列做计数统计（提示使用value_counts()）；
print(data['zwyx'].mean())

#取出所有列名
data_col = list(data.columns)
data_col.remove('zwyx')
data_col.remove('Id')
# data.loc[:,data_col]

dataFrame = pd.DataFrame([],columns = ['value'])
for i in data_col:
    d =  data[i].value_counts()
    new_d = pd.DataFrame(d)
    new_d.columns = ['value']
    dataFrame = pd.concat([dataFrame,new_d])
    
dataFrame
```

```python
#3，得到zwmc字段的唯一值列表；
list(data['zwmc'].unique())
```

```python
# 4，通过group函数，实现对于dd字段的分组，并按照城市计算每个城市的最大薪资，
#     使用折线图，显示Top10城市。（选做）


datas = data.groupby('dd')['zwyx'].max().sort_values(ascending=False).head(10)
# datas
datas.plot(label = '薪资')
plt.legend()
plt.title('城市月薪')
plt.xlabel('城市')
plt.ylabel('薪资/元')
plt.show()
```
