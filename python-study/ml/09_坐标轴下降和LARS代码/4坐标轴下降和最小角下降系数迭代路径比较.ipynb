# lasso_path和lars_path对比

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
#读取数据
data=pd.read_excel("boston.xlsx")
#data.info()
#data.describe().T
```

```python
#选择自变量和因变量
X=data.iloc[:,0:13]
Y=data.target
#数据标准化
ss=StandardScaler()
X= ss.fit_transform(X)
#lars方法系数路径
lars_alphas,_,lars_coefs= linear_model.lars_path(X, Y, method='lasso', verbose=True)
print("lars_alphas.shape：",lars_alphas.shape,"lars_coefs.shape:",lars_coefs.shape)
#coordinate descent 方法系数路径
lasso_alphas,lasso_coefs,_= linear_model.lasso_path(X, Y, verbose=True)
print("lasso_alphas.shape：",lasso_alphas.shape,"lasso_coefs.shape:",lasso_coefs.shape)
```

```python
#系数从0到迭代终止归一化
c_lars = np.sum(np.abs(lars_coefs), axis=0)
c_lars /= c_lars[-1]
c_lasso = np.sum(np.abs(lasso_coefs), axis=0)
c_lasso /= c_lasso[-1]
#呈现系数迭代路径图
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(c_lars, lars_coefs.T)
ymin, ymax = plt.ylim()
plt.vlines(c_lars, ymin, ymax, linestyle='dashed')
plt.xlabel('|coef| / max|coef|')
plt.ylabel('Coefficients')
plt.title('LARS Path')

plt.subplot(2,2,2)
plt.plot(c_lasso, lasso_coefs.T)
ymin, ymax = plt.ylim()
plt.vlines(c_lasso, ymin, ymax, linestyle='dashed')
plt.xlabel('|coef| / max|coef|')
plt.ylabel('Coefficients')
plt.title('LASSO-coordinate descent Path')
plt.axis('tight')
plt.show()
```

```python
#比较两种方法的路径:
plt.figure(figsize=(12,8))
plt.subplot(2,2,3)
L1=plt.plot(c_lasso, lasso_coefs.T)
L=plt.plot(c_lars, lars_coefs.T,linestyle="--")
plt.ylabel('coefficients')
plt.title('Lasso and lars')
plt.legend((L1[-1], L[-1]), ('Lasso', 'Lars'), loc='lower left')
plt.axis('tight')
plt.grid(b=True)
plt.show()
```
